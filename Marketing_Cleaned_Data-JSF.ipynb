```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Marketing_data.csv", sep=';')
```

```python
df = df.dropna()
df = df.drop_duplicates()
```

```python
df["Revenue"] = pd.to_numeric(df["Revenue"], errors="coerce")
df["Advertising_budget"] = pd.to_numeric(df["Advertising_budget"], errors="coerce")
df["Net_profit"] = pd.to_numeric(df["Net_profit"], errors="coerce")
```

```python
# KPI Summary
print("Total Revenue:", df["Revenue"].sum())
print("Total Profit:", df["Net_profit"].sum())
print("Average ROI:", df["ROI"].mean())
```

```python
# Revenue by Device
device_rev = df.groupby("User_device")["Revenue"].sum()
```

```python
plt.figure(figsize=(8,5))
device_rev.plot(kind="bar")
plt.title("Revenue by User Device")
plt.show()
```

```python
# ROI Distribution
plt.figure()
sns.histplot(df["ROI"], bins=20)
plt.title("ROI Distribution")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
